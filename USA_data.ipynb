```python
import numpy as np
import pandas as pd
```

```python
t1 = pd.read_excel('table1.xlsx')
```

```python
t1.shape
```

```python
# First, let's calculate the total number of missing values across the entire dataframe
total_missing_values = t1.isnull().sum().sum()

# Now, we calculate the total number of elements across the dataframe
total_elements = t1.size

# Finally, calculate the percentage of missing values
total_percentage_missing = (total_missing_values / total_elements) * 100
total_percentage_missing
```

```python
t1
```

```python
t1 = t1.dropna()

# Convert data types where applicable, using .loc to avoid SettingWithCopyWarning
for col in t1.columns[1:]:  # Assuming the first column is non-numeric (State names)
    t1.loc[:, col] = pd.to_numeric(t1[col], errors='coerce')
#t1 = t1.drop(t1.columns[-1], axis=1)
```

```python
t1
```

```python
# Convert the decimal fractions to percentages
t1.loc[:, 'Percent'] = t1.loc[:, 'Percent'] * 100
t1.loc[:, 'Percent.1'] = t1.loc[:, 'Percent.1'] * 100
t1.loc[:, 'Percent'] = t1.loc[:, 'Percent'].astype(str) + '%'
t1.loc[:, 'Percent.1'] = t1.loc[:, 'Percent.1'].astype(str) + '%'
```

```python
# Define the new column names with a list comprehension
new_column_names = ['State'] + [f'CO2_Emissions_{year}' for year in range(1970, 2022)] + ['CO2Change_Percent_1970_2021', 'CO2Change_Absolute_1970_2021', 'CO2Change_Percent_2019_2021', 'CO2Change_Absolute_2019_2021']

# Rename the columns
t1.columns = new_column_names
```

```python
t1
```

```python
t1.info()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Selecting the columns to scale, which in your case are the emission columns from 1970 to 2021
columns_to_scale = [col for col in t1.columns if 'CO2_Emissions' in col]

# Initialize the Min-Max Scaler
scaler = MinMaxScaler()

# Fit the scaler to the data and transform it with .loc to avoid SettingWithCopyWarning
t1.loc[:, columns_to_scale] = scaler.fit_transform(t1[columns_to_scale])
```

```python
t1
```

```python
t1.to_excel('t1_norm.xlsx',index=False)
```

```python
import matplotlib.pyplot as plt

columns = [col for col in t1.columns if 'CO2_Emissions' in col]

# Set the 'State' column as the index
data_selected = t1.set_index('State')[columns]

states_to_plot = ['California', 'Texas', 'New York']

# Create a line plot
plt.figure(figsize=(14, 7))
for state in states_to_plot:
    plt.plot(data_selected.columns, data_selected.loc[state], label=state)

plt.title('Trends in CO2 Emissions (1970-2021)')
plt.xlabel('Year')
plt.ylabel('CO2 Emissions (as % of national total)')
plt.xticks(rotation=45)
plt.legend(title='State')
plt.grid(True)
plt.tight_layout()
plt.show()
```
